```python
marcas = [['EXITO WOW', 'EXITO'],
          ['VIVAS', 'EXITO']]
```

```python
modelos_impresoras = pd.read_excel('data_base/Modelos_Impresoras.xlsx')

didetexco = pd.read_excel('data_base/Didetexco.xlsx')
didetexco.drop(['B&N', 'Color', 'Total'], axis=1, inplace=True)

dependencias = pd.read_excel('data_base/DEPENDENCIAS.xlsx', sheet_name='DEPENDENCIAS')
dependencias.dropna(subset=['CODIGO_DEPENDENCIA'], axis=0, inplace=True)

empleados_fondo = pd.read_excel('data_base/Usuarios Windows DA Presente.xlsx')

matriz = pd.read_excel('data_base/Matriz Instalacion ERC.xlsx', sheet_name='Matriz ERC')
matriz.SERIAL = matriz.SERIAL.astype(str)
matriz = matriz.rename(columns={'CIUDAD':'CIUDAD_MATRIZ', 'MARCA':'MARCA_MATRIZ'})
for j, marca in enumerate(marcas):
    matriz['MARCA_MATRIZ'].mask(matriz['MARCA_MATRIZ']==marcas[j][0], marcas[j][1], inplace=True)
matriz.ALMACEN = matriz.ALMACEN.astype(str)
matriz['COMPUSPAR'] = matriz.ALMACEN.map(lambda x : 'Si' if x.lower().find('compuspar') >= 0 else 'No')
matriz.dropna(subset=['NUMERO'], axis=0, inplace=True)
matriz = pd.merge(matriz, modelos_impresoras[['MODELO','Color_BN']], on='MODELO', how='left')
matriz = pd.merge(matriz, dependencias[['CODIGO_DEPENDENCIA', 'DEPENDENCIA', 'MARCA_ALMACEN', 'REGIONAL_DEFINITIVA', 'CIUDAD', 'DEPARTAMENTO', 'LATITUD', 'LONGITUD']], left_on='CodigoDependencia', right_on='CODIGO_DEPENDENCIA', how='left')
matriz.fillna(value=0, method=None, inplace=True)



facturacion = pd.read_excel('data_base/ALMACENES_billable.xlsx')
facturacion.dropna(subset=['Business Partner'], axis=0, inplace=True)
facturacion.dropna(subset=['Serial Number'], axis=0, inplace=True)
facturacion['Serial Number'] = facturacion['Serial Number'].astype(str)
facturacion['Invoice To'] = pd.to_datetime(facturacion['Invoice To'])
facturacion['MES'] = facturacion['Invoice To'].apply(lambda x : x.month)
facturacion['ANHO'] = facturacion['Invoice To'].apply(lambda x : x.year)
facturacion.loc[:, 'MES'] = facturacion.MES.map({1:'Enero', 2:'Febrero', 3:'Marzo', 4:'Abril', 5:'Mayo', 6:'Junio', 7:'Julio', 8:'Agosto', 9:'Septiembre', 10:'Octubre', 11:'Noviembre', 12:'Diciembre'})
facturacion = pd.merge(facturacion, matriz[['DIRECCION', 'SERIAL', 'AREA', 'CODIGO_DEPENDENCIA', 'DEPENDENCIA', 'MARCA_MATRIZ', 'MARCA_ALMACEN', 'REGIONAL_DEFINITIVA', 'CIUDAD', 'DEPARTAMENTO', 'LATITUD', 'LONGITUD']], left_on='Serial Number', right_on='SERIAL', how='left')
```

```python
data.drop(['SITE', 'DESTINATION','RELEASEHOSTNAME', 'PROFILE', 'CUSTOM2', 'CUSTOM3', 'SERIALNUMBER'], axis=1, inplace=True)
data.loc[:, 'FINALACTION'] = data.FINALACTION.map({'C':'Copia', 'D':'Cancelado', 'E':'Email', 'P':'Impresión', 'S':'Escaner'})
data.loc[:, 'PRINTJOBDUPLEX'] = data.PRINTJOBDUPLEX.map({'Y':'Si', 'N':'No'})
data.loc[:, 'PRINTJOBNAME'] = data.PRINTJOBNAME.apply(lambda x: impresion.limpiar(x))
data = pd.merge(data, matriz[['Color_BN', 'MODELO', 'DIRECCION', 'SERIAL', 'AREA', 'CODIGO_DEPENDENCIA', 'DEPENDENCIA', 'MARCA_MATRIZ', 'MARCA_ALMACEN', 'REGIONAL_DEFINITIVA', 'CIUDAD', 'DEPARTAMENTO', 'LATITUD', 'LONGITUD']], left_on='RELEASEIP', right_on='DIRECCION', how='left')
data = pd.merge(data, empleados_fondo, left_on='USERID', right_on='Nombre', how='left')
data.fillna(value=0, method=None, inplace=True)
data.loc[:, 'PRINTJOBCOLOR'] = data.PRINTJOBCOLOR.map({'Y':'Color', 'N':'B&N', 'U':'Indefinido'})
```

```python
data['MODELO'] = data['MODELO'].astype(str)
data['MARCA_MATRIZ'] = data['MARCA_MATRIZ'].astype(str)
data['USUARIO'] = data.apply(lambda x : impresion.nombre_usuario(x['USERID'], x['PRINTJOBNAME'], x['clase_archivo'], x['FINALACTION']), axis=1)
data['USUARIO'] = data['USUARIO'].astype(str)
data['TIPO_USUARIO'] = data.apply(lambda x : impresion.tipo_usuario(x['USUARIO'], x['MODELO'], x['MARCA_MATRIZ']), axis=1)
data['COLOR_TRABAJO'] = data.apply(lambda x : impresion.color_impresion(x['PRINTJOBCOLOR'], x['Color_BN']), axis=1)

data.loc[:,'SUBMITDATE'] = pd.to_datetime(data['SUBMITDATE'])
data['MES'] = data.SUBMITDATE.apply(lambda x : x.month)
data.loc[:, 'MES'] = data.MES.map({1:'Enero', 2:'Febrero', 3:'Marzo', 4:'Abril', 5:'Mayo', 6:'Junio', 7:'Julio', 8:'Agosto', 9:'Septiembre', 10:'Octubre', 11:'Noviembre', 12:'Diciembre'})
data['ANHO'] = data.SUBMITDATE.apply(lambda x : x.year)
#Limpiar de comas y punto y comas los nombres de los trabajos
#especificar si el usuario pertenece a fondo presente
data['FONDO_PRESENTE'] = data.apply(lambda x : impresion.clas_usuario(x['USUARIO'], x['Nombre']), axis=1)
#data.drop(['Nombre'], axis=1, inplace=True)

data_copia_impresion = data[data.FINALACTION.isin(['Copia', 'Impresión'])]
#Eliminar los registros mayores a 500, q no corresponden
#automatizar
#indices_eliminar = [27651, 74567, 80470, 277579, 374505, 792707, 795245, 1007197, 1049549, 1390506, 1397928, 1489146, 1497407, 1497429, 1854283, 1859802, 1906541, 1913826, 1943598, 1963048, 2097737, 2168217, 2169277, 2202140, 2254373, 2254635, 2304136, 2414180]
#data_copia_impresion.drop(index=indices_eliminar, inplace=True)
```

```python
#Filtrar los usuarios que no sean de fondo presente
data_corporativo = data_copia_impresion[(data_copia_impresion['MARCA_MATRIZ'] == 'CORPORATIVO') & (data_copia_impresion['FONDO_PRESENTE'] == 'Administrativo')]
data_corporativo = data_corporativo[['USUARIO', 'SERIAL', 'AREA', 'DEPENDENCIA', 'COLOR_TRABAJO', 'PRINTJOBNAME', 'MES', 'ANHO', 'NUMPAGES']]
data_corporativo = data_corporativo.groupby(['USUARIO', 'SERIAL', 'AREA', 'DEPENDENCIA', 'COLOR_TRABAJO', 'PRINTJOBNAME', 'MES', 'ANHO']).agg({'NUMPAGES':'sum'}).reset_index()
data_corporativo.columns = ['USUARIO', 'SERIAL', 'AREA', 'DEPENDENCIA', 'COLOR_TRABAJO', 'PRINTJOBNAME', 'MES', 'ANHO', 'NUMPAGES']
```

```python
#Trabajos de usuarios corporativos bn y color
usuarios_corporativos_bn = data_corporativo[data_corporativo.COLOR_TRABAJO == 'B&N'].groupby(['USUARIO', 'AREA', 'DEPENDENCIA', 'MES', 'ANHO']).agg({'NUMPAGES':'sum'}).reset_index().sort_values(by='NUMPAGES', ascending=False)
usuarios_corporativos_color = data_corporativo[data_corporativo.COLOR_TRABAJO == 'Color'].groupby(['USUARIO', 'AREA', 'DEPENDENCIA', 'MES', 'ANHO']).agg({'NUMPAGES':'sum'}).reset_index().sort_values(by='NUMPAGES', ascending=False)
```

```python
#top 10 corporativo B&N
top_corporativo_bn = data_corporativo[data_corporativo.COLOR_TRABAJO == 'B&N'].groupby(['USUARIO']).agg({'NUMPAGES':'sum'}).reset_index()
top_corporativo_bn.sort_values(by='NUMPAGES', ascending=False, inplace=True)
top_10_usuarios_corporativo_bn = top_corporativo_bn.head(15)
```

```python
#top 10 corporativo Color
top_corporativo_color = data_corporativo[data_corporativo.COLOR_TRABAJO == 'Color'].groupby(['USUARIO']).agg({'NUMPAGES':'sum'}).reset_index()
top_corporativo_color.sort_values(by='NUMPAGES', ascending=False, inplace=True)
top_10_usuarios_corporativo_color = top_corporativo_color.head(15)
```

```python
#top 5 corporativo B&N
#Cambiar al usuario definitivo
top_corporativo_bn = data_corporativo[data_corporativo.COLOR_TRABAJO == 'B&N'].groupby(['USUARIO']).agg({'NUMPAGES':'sum'}).reset_index()
top_corporativo_bn.sort_values(by='NUMPAGES', ascending=False, inplace=True)
usuarios_corporativo_bn = top_corporativo_bn.head(5)['USUARIO']
top_5_corporativo_bn = pd.DataFrame(columns=['USUARIO', 'SERIAL', 'AREA', 'DEPENDENCIA', 'COLOR_TRABAJO', 'PRINTJOBNAME', 'NUMPAGES'])

for usuario in usuarios_corporativo_bn:
    df_temporal = data_corporativo[(data_corporativo['USUARIO'] == usuario) & (data_corporativo['COLOR_TRABAJO'] == 'B&N')]
    df_temporal.sort_values(by='NUMPAGES', ascending=False, inplace=True)
    df_temporal = df_temporal.head(5)
    top_5_corporativo_bn = top_5_corporativo_bn.append(df_temporal)
```

```python
#top 5 corporativo Color
#Cambiar al usuario definitivo
top_corporativo_color = data_corporativo[data_corporativo.COLOR_TRABAJO == 'Color'].groupby(['USUARIO']).agg({'NUMPAGES':'sum'}).reset_index()
top_corporativo_color.sort_values(by='NUMPAGES', ascending=False, inplace=True)
usuarios_corporativo_color = top_corporativo_color.head(5)['USUARIO']
top_5_corporativo_color = pd.DataFrame(columns=['USUARIO', 'SERIAL', 'AREA', 'DEPENDENCIA', 'COLOR_TRABAJO', 'PRINTJOBNAME', 'NUMPAGES'])

for usuario in usuarios_corporativo_color:
    df_temporal = pd.DataFrame()
    df_temporal = data_corporativo[(data_corporativo['USUARIO'] == usuario) & (data_corporativo['COLOR_TRABAJO'] == 'Color')]
    df_temporal.sort_values(by='NUMPAGES', ascending=False, inplace=True)
    df_temporal = df_temporal.head(5)
    top_5_corporativo_color = top_5_corporativo_color.append(df_temporal)
```
