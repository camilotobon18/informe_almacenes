```python
import pandas as pd
import re
#-⁻- coding: UTF-8 -*-
```

```python
class Impresion():
    def limpiar(self, texto):
        texto = str(texto)
        texto = texto.lower()
        texto = re.sub(r'_', ' ', texto)
        texto = re.sub(r'-', ' ', texto)
        texto = re.sub(r'  ', ' ', texto)
        texto = re.sub(r'^\s|\s$', '', texto)
        texto = re.sub(r'[àá]', 'a', texto)
        texto = re.sub(r'[èé]', 'e', texto)
        texto = re.sub(r'[íì]', 'i', texto)
        texto = re.sub(r'[òó]', 'o', texto)
        texto = re.sub(r'[úù]', 'u', texto)
        texto = re.sub(r'[;,]', '_', texto)
        return texto
    

    def tipo_usuario(self, usuario, modelo, marca):
        usuario = usuario.lower()
        modelo = modelo.lower()
        marca = marca.lower()
        tipo = ""
        if usuario.find('pop') >= 0:
            tipo = "POP"
        elif usuario.find('marcaci') >= 0:
            tipo = "Marcación"
        elif usuario.find('decoraci') >= 0:
            tipo = "Decoración"
        elif (modelo == 'cs923de') or ((modelo == 'MS823dn') and (marca.find('exito') >= 0)):
            tipo = "POP"
        elif modelo == 'ms911de':
            tipo = "POP"
        elif (modelo == 'cs622de') and (marca.find('carulla') >= 0):
            tipo = "POP"
        else:
            tipo = "OTROS USUARIOS"  

        return tipo
    
    #condicional, color verdadero de la impresión
    def color_impresion(self, color_trabajo, tipo_impresora):
        nuevo_color_impresion = ''
        if (color_trabajo == 'Color') and (tipo_impresora == 'B&N'):
            nuevo_color_impresion = 'B&N'
        else:
            nuevo_color_impresion = color_trabajo

        return nuevo_color_impresion
    
    #Nombre real del archivo
    def nombre_usuario(self, usuario, nombre_trabajo, tipo_archivo, tipo_servicio):
        if tipo_archivo.find('Sinco') >= 0:
            nom_usuario = 'Impresión Sinco'
        elif nombre_trabajo == 'genericnetworkjob':
            nom_usuario = 'Impresión Sinco'
        elif nombre_trabajo == "":
            if tipo_servicio == "Copia":
                nom_usuario = "Copia"
            else:
                nom_usuario = "Sin usuario"
        elif (tipo_servicio == "Copia") and (usuario == ""):
            nom_usuario = "Copia"
        else:
            nom_usuario = usuario

        return nom_usuario
    
    #Funcion para hallar el promedio de las impresiones
    def imp_prom(self, dividendo, divisor):
        if dividendo > 0:
            resultado = dividendo / divisor
        else:
            resultado = 0
        return resultado
    
    #Hallar si el usuario es o no administrativo
    def clas_usuario(self, usuario, nombre):
        usuario = str(usuario)

        no_administrativo = re.compile(r'5017|137|36|501')
        if no_administrativo.search(usuario):
            resultado = "No_administrativo"
        else:
            if nombre == 0:
                resultado = 'Administrativo'
            else:
                resultado = 'Fondo_presente'

        return resultado
```
